```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from GRN_aux_5 import get_result_static, get_results
```

```python
x = []
A = []
B = []
C = []
D = []
E = []
with open(f'GRN5.txt', 'r', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='\'')
    for row in reader:
       x.append(float(row[0]))
       A.append(float(row[1]))
       B.append(float(row[2]))
       C.append(float(row[3]))
       D.append(float(row[4]))
       E.append(float(row[5]))

print(x)
print(A)
print(B)
print(C)
print(D)
print(E)
```

```python

fig,ax=plt.subplots(1,1)
ax.plot(x, A, label='A')
ax.plot(x, B, label='B')
ax.plot(x, C, label='C')
ax.plot(x, D, label='D')
ax.plot(x, E, label='E')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()

print(max(A))
print(max(B))
print(max(C))
print(max(D))
print(max(E))
```
