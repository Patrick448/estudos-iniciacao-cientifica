```python
import matplotlib.pyplot as plt
import pandas as pd
import string

def plot_series(path, title):
    df = pd.read_csv(path, index_col=0, sep=",", header=None)
    series =[float(x) for x in  str(df.iloc[0][1]).split(",")]
    pd_series = pd.Series(series)
    pd_series.plot(title=title)
    plt.show()
    
def plot_series_from_seed(path, title, seed):
    df = pd.read_csv(path, index_col=0, sep=",", header=None)
    series =[float(x) for x in  str(df.loc[seed][1]).split(",")]
    #series =[float(x) for x in  str(df.iloc[0][1]).split(",")]
    pd_series = pd.Series(series)
    pd_series.plot(title=title)
    plt.show()
    
def plot_all_series(path_prefix, algorirthms):
    for alg in algorirthms:
        plot_series(path_prefix+'-'+alg+'.csv', alg)
        
def get_best_individual(model, experiment_suffix, algorithm):
    df = pd.read_csv(f'../results/{model}-{experiment_suffix}/output-{model}-{algorithm}.csv', index_col=0, sep=",")
    best = df['eval'].idxmin()
    
    return best

#def get  
```

```python
best_seed = get_best_individual('abcd', '1050000-norm_v2', 'es-ni')
plot_series_from_seed('../results/abcd-1050000-norm_v2/progress-abcd-es-ni.csv', 'ES-NI', best_seed)

best_seed = get_best_individual('abcd', '1050000-norm_v2', 'cmaes')
plot_series_from_seed('../results/abcd-1050000-norm_v2/progress-abcd-cmaes.csv', 'CMAES', best_seed)
```
