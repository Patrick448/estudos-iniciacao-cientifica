```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd

def get_comparison_df(experiment_folder, experiment_prefix, alg_names):
    algs_dfs = []
    comparison_dict = {}
    for alg_name in alg_names:
        algs_dfs.append(pd.read_csv(f'../results/{experiment_folder}/output-{experiment_prefix}-{alg_name}.csv', index_col=False))
    
    for i,alg_name in enumerate(alg_names):
        comparison_dict[alg_name] = algs_dfs[i]['eval']
    
    return pd.DataFrame(comparison_dict)
```

# Comparação

Comparação entre os seguintes métodos:
<br>CMAES
<br>(15+105)-ES isotrópico
<br>(15+105)-ES não isotrópico
<br>DE
<br>SADE


# GRN com 5 variáveis - 1050000 chamadas

```python
get_comparison_df('grn5new-1050000-norm_v2','grn5new', ['cmaes', 'de', 'sade', 'es-i', 'es-ni']).boxplot(column=['cmaes', 'de', 'sade', 'es-i', 'es-ni'])
plt.show()
```
