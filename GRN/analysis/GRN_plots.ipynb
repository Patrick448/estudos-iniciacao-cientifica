```python
def load_grn_dataframe(filename, col_names):
   data = pd.read_csv(filename, header=None, sep=" ")
   data.columns = col_names
  # data.head()
   return data

def load_results_dataframe(results, time_col, col_names):
    result_df = pd.DataFrame(results)
    result_df = result_df.transpose()
    result_df.insert(0, 't', time_col)
    result_df.columns = col_names
    return result_df

def plot_compare_each(expected, obtained, x_label, y_labels, obtained_start_at=None):
    s = obtained_start_at if obtained_start_at else 0
    
    for y_label in y_labels:
        plt.figure(figsize=(8,3))
        plt.plot(expected[x_label], expected[y_label], label='Esperado', marker='.') 
        plt.plot(obtained[x_label][s:], obtained[y_label][s:], label='Obtido', marker='.')
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.legend()
        
        plt.show()
        
```
