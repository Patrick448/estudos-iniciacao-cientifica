```python
import matplotlib.pyplot as plt
import pandas as pd
import string

def plot_comparison(actual_series, predicted_series, train_pred=None):
    
    for col in predicted_series.columns:
        comparison_df = pd.DataFrame({'actual':actual_series[col], 'predicted':predicted_series[col], 'tr.pred':train_pred[col] if train_pred is not None else ''})
        comparison_df.plot(title=col, marker='o')
        plt.show()
```

```python
pred = pd.read_csv('../results/grn10new2-1050000/grn10new2_best_test.txt', index_col=0, sep="\t", header=None, names=list(string.ascii_uppercase[:10]))
pred.plot(marker='o')

actual_series = pd.read_csv('GRN10_DATA.txt', index_col=0, sep="\t", header=None, names=list(string.ascii_uppercase[:11]))

train_pred = pd.read_csv('../results/grn10new2-1050000/grn10new2_best_train.txt', index_col=0, sep="\t", header=None, names=list(string.ascii_uppercase[:10]))

actual_series.drop(columns=['K'], inplace=True)
actual_series[51.4286:].plot(marker='o')

plt.show()

plot_comparison(actual_series, pred, train_pred)
  
```
