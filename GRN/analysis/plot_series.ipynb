```python
import matplotlib.pyplot as plt
import pandas as pd
import string

def plot_comparison(actual_series, predicted_series, train_pred=None, save_to=None):
    
    for col in predicted_series.columns:
        comparison_df = pd.DataFrame({'ac':actual_series[col], 'pr':predicted_series[col], 'tr':train_pred[col] if train_pred is not None else ''})
        ax = comparison_df.plot(title=col, marker='o', figsize=(8, 2))
        #ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
        if save_to:
            plt.savefig(f'{save_to}/output_{col}.png')
        plt.show()
```

## Modelo de 5 vari√°veis

```python
actual_series_grn5new= pd.read_csv('GRN5_DATA.txt', index_col=0, sep="\t", header=None,names=list(string.ascii_uppercase[:6]))
actual_series_grn5new.drop(columns=['F'], inplace=True)
actual_series_grn5new[51.428600:].plot(marker='o', title="Actual")

pred_grn5new = pd.read_csv('../results/grn5new-1050000/grn5new_best_test.txt', index_col=0, sep="\t", header=None, names=list(string.ascii_uppercase[:5]),)
pred_grn5new.plot(marker='o', title="Predicted")

train_grn5new = pd.read_csv('../results/grn5new-1050000/grn5new_best_train.txt', index_col=0, sep="\t", header=None, names=list(string.ascii_uppercase[:5]),)

plot_comparison(actual_series_grn5new, pred_grn5new, train_grn5new, save_to='graficos/grn5')
```
