```python
from ES import ESAlgorithm
import matplotlib.pyplot as plt
from py_expression_eval import Parser

```

```python
parser = Parser()
expr = "sin(x+y) + (x-y)^2 - 1.5*x + 2.5*y +1"
def allowed_interval(**kwargs):
    if -1.5 <= kwargs['x'] <= 4 and -3.0 <= kwargs['y'] <= 4:
        return True

    return False


def evaluate(ind_dict):

    if allowed_interval(**ind_dict):
        print(f"ind_dict: {ind_dict}        eval: {parser.parse(expr).evaluate(ind_dict)}")
        return parser.parse(expr).evaluate(ind_dict)

    return float('inf')

alg = ESAlgorithm()
alg.set_evaluation_function(evaluate)
result, best = alg.run_one_plus_one({'x':0, 'y':0}, iterations=50, seed=0)
print('\n'.join([str(ind) for ind in result]))
eval_hist = [ind.get_evaluation() for ind in result]
print(result[len(result)-1])
print(evaluate({'x': -4.998190785187728, 'y': 2.1481530531545396}))

#fig = plt.figure()  # an empty figure with no Axes
#fig, axs = plt.subplots(1, 1)
print(eval_hist)
plt.plot(eval_hist)
plt.show()
plt.clf()
plt.close()
#axs.plot(eval_hist)
```
