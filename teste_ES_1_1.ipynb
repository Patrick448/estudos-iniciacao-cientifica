# Testes de otimização com ES(1+1)

## Testes com funções com mínimo local global

```python
alg = ESAlgorithm()
alg.set_evaluation_expression("sin(x+y) + (x-y)^2 - 1.5*x + 2.5*y +1")
alg.set_variable_bounds('x', 4, True, -1.5, True)
alg.set_variable_bounds('y', 4, True, -3, True)

history  = alg.one_plus_one_ES( sigma=0.5, c=0.817, n=10, iter=250, seed=0)
```

```python

x_history = [item['dim']['x'] for item in history]
y_history = [item['dim']['y'] for item in history]
obj_func_hist = [item['eval'] for item in history]
print(history[-2:-1])
```

```python
#print(history)
fig = plt.figure()  # an empty figure with no Axes
fig, axs = plt.subplots(3, 1)
axs[0].plot(x_history)
axs[1].plot(y_history)
axs[2].plot(obj_func_hist)
```

```python
def allowed_interval(**kwargs):
    if -10 <= kwargs['x'] <= 10 and -10 <= kwargs['y'] <= 10:
        return True
    return False

history, last_ind, obj_func_hist  = alg1plus1ES(expr="(x + 2*y - 7)^2 + (2*x + y - 5)^2", validator=allowed_interval, sigma=0.5, c=0.817, n=10, iter=250, seed=0)
x_history = [item['x'] for item in history]
y_history = [item['y'] for item in history]

fig = plt.figure()  # an empty figure with no Axes
fig, axs = plt.subplots(3, 1)

axs[0].plot(x_history)
axs[1].plot(y_history)
axs[2].plot(obj_func_hist)

print(last_ind)
```

## Funções com multiplos mínimos locais

```python
def allowed_interval(**kwargs):
    if -10 <= kwargs['x'] <= 10 and -10 <= kwargs['y'] <= 10:
        return True
    return False

history, last_ind, obj_func_hist  = alg1plus1ES(expr="- abs (sin(x)*cos(y)*exp(abs(1 - ((sqrt(x^2 +y^2))/PI))))", validator=allowed_interval, sigma=0.5, c=0.817, n=10, iter=250, seed=0)
x_history = [item['x'] for item in history]
y_history = [item['y'] for item in history]

fig = plt.figure()  # an empty figure with no Axes
fig, axs = plt.subplots(3, 1)

axs[0].plot(x_history)
axs[1].plot(y_history)
axs[2].plot(obj_func_hist)

print(last_ind)
```
