# Testes de otimização com ES(1+1)

## Testes com funções com mínimo local global

```python
alg = ESAlgorithm()
alg.set_evaluation_expression("sin(x+y) + (x-y)^2 - 1.5*x + 2.5*y +1")
alg.set_variable_bounds('x', 4, True, -1.5, True)
alg.set_variable_bounds('y', 4, True, -3, True)

history  = alg.one_plus_one_ES( sigma=0.5, c=0.817, n=10, iter=250, seed=0)
```

```python

x_history = [item['dim']['x'] for item in history]
y_history = [item['dim']['y'] for item in history]
obj_func_hist = [item['eval'] for item in history]
print(history[-2:-1])
```

```python
#print(history)
fig, axs = plt.subplots(3, 1)
axs[0].plot(x_history)
axs[1].plot(y_history)
axs[2].plot(obj_func_hist)
```

```python
alg = ESAlgorithm()
alg.set_evaluation_expression("(x + 2*y - 7)^2 + (2*x + y - 5)^2")
alg.set_variable_bounds('x', 10, True, -10, True)
alg.set_variable_bounds('y', 10, True, -10, True)

history = alg.one_plus_one_ES(sigma=0.5, c=0.817, n=10, iter=250, seed=0)

x_history = [item['dim']['x'] for item in history]
y_history = [item['dim']['y'] for item in history]
obj_func_hist = [item['eval'] for item in history]
print(history[-2:-1])
#print(history)
fig, axs = plt.subplots(3, 1)
axs[0].plot(x_history)
axs[1].plot(y_history)
axs[2].plot(obj_func_hist)
```

## Funções com multiplos mínimos locais

```python
alg = ESAlgorithm()
alg.set_evaluation_expression("- abs (sin(x)*cos(y)*exp(abs(1 - ((sqrt(x^2 +y^2))/PI))))")
alg.set_variable_bounds('x', 10, True, -10, True)
alg.set_variable_bounds('y', 10, True, -10, True)

history = alg.one_plus_one_ES(sigma=0.5, c=0.817, n=10, iter=250, seed=0)

x_history = [item['dim']['x'] for item in history]
y_history = [item['dim']['y'] for item in history]
obj_func_hist = [item['eval'] for item in history]
print(history[-2:-1])
#print(history)
fig, axs = plt.subplots(3, 1)
axs[0].plot(x_history)
axs[1].plot(y_history)
axs[2].plot(obj_func_hist)
```
